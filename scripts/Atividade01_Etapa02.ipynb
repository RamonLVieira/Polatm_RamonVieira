## Numpy arrays

```python
import numpy as np

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2)

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data))

# Print de uma amostra dos dados
print(data[0:10])
print(data[0:10].shape)
#Este código cria uma matriz de 1000 linhas e 2 colunas contendo números aleatórios entre 0 e 1 usando NumPy. Em seguida, imprime o formato e o tamanho dessa matriz, além de uma amostra das primeiras 10 linhas.
```

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[0:10])
#O código define valores máximos e mínimos para dados, realiza uma transformação linear nesses dados e imprime os primeiros 10 elementos do resultado.
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3)
ax[0].plot(data)
ax[1].plot(data[:,0])
ax[2].plot(data[:,1])
print(type(data))
#Neste código, estamos usando a biblioteca matplotlib.pyplot para criar um gráfico com três subplots empilhados verticalmente.
#fig, ax = plt.subplots(3): Criamos uma figura (fig) e uma matriz de eixos (ax) com 3 linhas de subplots.
#ax[0].plot(data): Plotamos os dados (data) no primeiro subplot.
#ax[1].plot(data[:,0]): Plotamos os dados da primeira coluna de data no segundo subplot.
#ax[2].plot(data[:,1]): Plotamos os dados da segunda coluna de data no terceiro subplot.
```

```python
# Criando uma função
def noiseCreation(path, year):
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # Monstrando um dataframe
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4)))
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red')
    df.to_csv(path+'/noise_'+str(year)+'.csv')
    
    return df
```
