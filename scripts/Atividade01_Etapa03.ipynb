## Contextualização da empresa:

O Portobello Grupo (PBG S.A.) é uma empresa brasileira, com atuação global, que possui a maior rede de revestimentos cerâmicos do país.
De acordo com o Relatório de Sustentabilidade da empresa, em 2022 o Portobello Grupo foi responsável pela fabricação de 716.952 toneladas de revestimento cerâmico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a produção anual em toneladas
producao_anual = 716952

print(f"A produção Anual do Portobello Gupo é de {producao_anual} toneladas de material cerâmico")
```

```python
# Definindo o número de dias em um ano (365)
dias_no_ano = 360 # Ajustado para que seja divisível por 12
```

```python
# Gerando produção diária aleatória para um ano
producao_diaria = np.random.randint(1000, 3000, dias_no_ano)

# Calculando a produção mensal
producao_mensal = np.split(producao_diaria, 12)

# Calculando a produção total mensal
producao_total_mensal = [np.sum(mes) for mes in producao_mensal]

# Plotando o gráfico
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
plt.figure(figsize=(10, 6))
plt.bar(meses, producao_total_mensal, color='skyblue')
plt.title('Produção Mensal de Material Cerâmico - Portobello Grupo (2022)')
plt.xlabel('Mês')
plt.ylabel('Produção (toneladas)')
plt.grid(True)
plt.show()
```
