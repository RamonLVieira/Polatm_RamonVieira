```python
# Criando uma lista em python
list = [0,'eu amo polatm', 2.5]
print(list)

# Neste trecho é feita a adição entre duas strings, sendo que list[1] é o segundo indice do vetor list
print('variavel: '+ list[1])

# Neste trecho é feita a adição do primeiro indice da variavel list (list[0]) - lembrar que a indexação em python inicia-se em 0
print(list[0] + 1)
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma matriz de zeros utilizando  np.zeros
a = np.zeros((5,5))
print(a)
print('indexação numpy')
print(a[0,0])
```

```python
# Cruiando números aleatórios e mostrando a dimensão da matriz
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
# Criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 # valor mínimo
vmax = 100 # valor máximo

c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
# Plotando apenas a primeira linha

x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
plt.title('Gráfico com Valores Interiros no Eixo X')
plt.xticks(x) # Definindo os valores do eixo x
```

```python
# Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# Exportar a gigura
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\',"/")+'/figures/'
print(figFolder)
fig.savefig(figFolder+'Figura01.png')
```
